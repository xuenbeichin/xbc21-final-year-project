```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_1samp
from scipy.sparse import coo_matrix
from mne.stats import permutation_cluster_1samp_test
from nilearn import datasets, surface, plotting
from nilearn.image import new_img_like, get_data


base_dir = "/Users/xuenbei/Desktop/finalyearproject/PDCM/fitted_data"
subject_ids = [1, 2, 3, 4, 6, 9, 10, 11, 12, 13, 15, 17, 18, 19, 20]
roi_list = list(range(100))

spdcm_parameters = [
    "mtt", "tau", "alpha", "E0",
    "alpha_v", "beta_v", "alpha_e", "beta_e",
    "sigma", "lamb", 'mu', "phi", "varphi", "chi"
]

n_permutations = 5000
alpha_threshold = 0.05
n_rois_total = 100


network_groups = [
    list(range(0, 9)), list(range(50, 58)), list(range(9, 15)), list(range(58, 66)),
    list(range(66, 72)), list(range(23, 30)), list(range(73, 77)), list(range(30, 33)),
    list(range(78, 80)), list(range(33, 37)), list(range(80, 89)), list(range(37, 50)),
    list(range(89, 100))
]
adj_matrix = np.zeros((n_rois_total, n_rois_total), dtype=int)
for group in network_groups:
    for i in group:
        for j in group:
            if i != j:
                adj_matrix[i, j] = 1
adjacency_sparse = coo_matrix(adj_matrix)


parameter_data = {param: {} for param in spdcm_parameters}
missing_files_log = {param: [] for param in spdcm_parameters}
sig_matrix = pd.DataFrame(0, index=spdcm_parameters, columns=range(n_rois_total))
results_df_rows = []


for subject in subject_ids:
    subject_key = f"sub-{subject:03d}"
    for roi in roi_list:
        path_plcb = os.path.join(base_dir, subject_key, f"roi-{roi}-PLCB_results.json")
        path_lsd = os.path.join(base_dir, subject_key, f"roi-{roi}-LSD_results.json")

        try:
            if not os.path.exists(path_plcb) or not os.path.exists(path_lsd):
                for param in spdcm_parameters:
                    missing_files_log[param].append((subject, roi, 'file missing'))
                continue

            with open(path_plcb) as f: data_plcb = json.load(f)
            with open(path_lsd) as f: data_lsd = json.load(f)

            params_plcb = data_plcb.get("final_parameters", {})
            params_lsd = data_lsd.get("final_parameters", {})

            for param in spdcm_parameters:
                if param == "A_real":
                    val_plcb = params_plcb.get("A", [[[None, None]]])[0][0][0]
                    val_lsd = params_lsd.get("A", [[[None, None]]])[0][0][0]
                elif param == "A_imag":
                    val_plcb = params_plcb.get("A", [[[None, None]]])[0][0][1]
                    val_lsd = params_lsd.get("A", [[[None, None]]])[0][0][1]
                else:
                    val_plcb = params_plcb.get(param, None)
                    val_lsd = params_lsd.get(param, None)

                if val_plcb is None or val_lsd is None:
                    missing_files_log[param].append((subject, roi, 'parameter missing'))
                    continue

                subj_data = parameter_data[param].setdefault(subject, {})
                subj_data[roi] = val_lsd - val_plcb

        except Exception as e:
            for param in spdcm_parameters:
                missing_files_log[param].append((subject, roi, f'error: {str(e)}'))


results = []
cluster_table_df = pd.DataFrame()

for param, subj_roi_dict in parameter_data.items():
    print(f"\n Testing parameter: {param} 
    subjects_with_data = sorted(subj_roi_dict.keys())
    n_subjects = len(subjects_with_data)

    diff_array = np.full((n_subjects, n_rois_total), np.nan)
    for i, subject in enumerate(subjects_with_data):
        for roi, value in subj_roi_dict[subject].items():
            diff_array[i, roi] = value

    missing_counts = np.isnan(diff_array).sum(axis=0)
    retained_rois = [i for i in range(n_rois_total) if missing_counts[i] == 0]
    if len(retained_rois) < 2:
        print("Not enough ROIs. Skipping.")
        continue

    diff_array_clean = diff_array[:, retained_rois]
    adj_matrix_clean = adj_matrix[np.ix_(retained_rois, retained_rois)]
    adjacency_sparse_clean = coo_matrix(adj_matrix_clean).tocsr()

    t_vals, p_vals_uncorrected = ttest_1samp(diff_array_clean, 0.0, axis=0)
    for i, (p_val, t_val) in enumerate(zip(p_vals_uncorrected, t_vals)):
        roi = retained_rois[i]
        results_df_rows.append({
            "parameter": param,
            "roi": roi,
            "p_value": p_val,
            "t_value": t_val
        })
        print(f"Parameter: {param}, ROI: {roi}, t = {t_val:.3f}, p = {p_val:.5f}")

    t_obs, clusters, cluster_pv, _ = permutation_cluster_1samp_test(
        diff_array_clean, threshold=None, tail=0,
        adjacency=adjacency_sparse_clean,
        n_permutations=n_permutations,
        out_type="indices", seed=42
    )

    for i, (roi_idxs,) in enumerate(clusters):
        p_val = cluster_pv[i]
        global_rois = [retained_rois[idx] for idx in roi_idxs]
        cluster_table_df = pd.concat([
            cluster_table_df,
            pd.DataFrame({
                "parameter": [param],
                "cluster_id": [i],
                "p_value": [p_val],
                "rois": [global_rois]
            })
        ], ignore_index=True)

        if p_val < alpha_threshold:
            for r in global_rois:
                sig_matrix.loc[param, r] = 1

        results.append({
        "parameter": param,
        "n_subjects": n_subjects,
        "n_rois": len(retained_rois),
        "n_clusters": len(clusters),
        "n_sig_clusters": sum(p < alpha_threshold for p in cluster_pv)
    })


results_df = pd.DataFrame(results)
print("\nSummary:")
print(results_df)

if sig_matrix.values.sum() > 0:
    plt.figure(figsize=(14, 6))
    sns.heatmap(sig_matrix, cmap="Reds", linewidths=0.5, linecolor="gray", cbar=False)
    plt.title("Significant Parameter × ROI Clusters (p < 0.05)")
    plt.xlabel("ROI")
    plt.ylabel("Parameter")
    plt.tight_layout()
    plt.show()
else:
    print("\nNo significant clusters found.")

print("\nSignificant Clusters:")
print(cluster_table_df[cluster_table_df["p_value"] < alpha_threshold])

# === SURFACE PLOTTING ===
schaefer = datasets.fetch_atlas_schaefer_2018(n_rois=100, yeo_networks=17, resolution_mm=1)
atlas_img = schaefer['maps']
atlas_labels = [label.decode("utf-8") for label in schaefer['labels']]
fsaverage = datasets.fetch_surf_fsaverage()
roi_to_hemi = ["left" if "LH" in label else "right" for label in atlas_labels]
roi_pval_df = pd.DataFrame(results_df_rows)
param_display_names = {'mtt': 'tau', 'tau': 'tau_1', 'lamb': 'lambda'}

corrected_logp_map = {}

for _, row in cluster_table_df.iterrows():
    if row["p_value"] < alpha_threshold:
        param = row["parameter"]
        logp = -np.log10(row["p_value"])
        for roi in row["rois"]:
            corrected_logp_map[(param, roi)] = logp

for param in roi_pval_df['parameter'].unique():
    subset = roi_pval_df[roi_pval_df['parameter'] == param]
    roi_pvals = dict(zip(subset['roi'], subset['p_value']))
    pval_map = np.zeros(100)
    for roi in range(100):
        key = (param, roi)
        if key in corrected_logp_map:
            pval_map[roi] = corrected_logp_map[key]

    atlas_data = get_data(atlas_img)
    stat_data = np.zeros_like(atlas_data)
    for roi_idx in range(100):
        stat_data[atlas_data == roi_idx + 1] = pval_map[roi_idx]
    stat_img = new_img_like(atlas_img, stat_data)
    for hemi in ['left', 'right']:
        hemi_rois = [i for i in range(100) if roi_to_hemi[i] == hemi and pval_map[i] > 0]
        if not hemi_rois:
            continue
        pial = fsaverage.pial_left if hemi == 'left' else fsaverage.pial_right
        infl = fsaverage.infl_left if hemi == 'left' else fsaverage.infl_right
        sulc = fsaverage.sulc_left if hemi == 'left' else fsaverage.sulc_right
        texture = surface.vol_to_surf(stat_img, pial)
        fig = plt.figure(figsize=(8, 6))
        display_param = param_display_names.get(param, param)
        fig = plotting.plot_surf_stat_map(
            surf_mesh=infl,
            stat_map=texture,
            hemi=hemi,
            bg_map=sulc,
            threshold=1.3,
            colorbar=True,
            cmap="Reds",
            title=None  
        )
        

        fig.axes[0].set_title(
            f"Significant ROIs for {display_param} ({hemi.capitalize()} Hemisphere)",
            fontsize=20  
        )


        if hasattr(fig, 'axes') and len(fig.axes) > 1:
            fig.axes[-1].set_ylabel("-log10(p-value)", fontsize=18)

plt.show()


significant_roi_rows = []


t_obs_per_param_roi = {}

for param, subj_roi_dict in parameter_data.items():
    subjects_with_data = sorted(subj_roi_dict.keys())
    n_subjects = len(subjects_with_data)

    diff_array = np.full((n_subjects, n_rois_total), np.nan)
    for i, subject in enumerate(subjects_with_data):
        for roi, value in subj_roi_dict[subject].items():
            diff_array[i, roi] = value

    missing_counts = np.isnan(diff_array).sum(axis=0)
    retained_rois = [i for i in range(n_rois_total) if missing_counts[i] == 0]
    if len(retained_rois) < 2:
        continue

    diff_array_clean = diff_array[:, retained_rois]
    t_obs, clusters, cluster_pv, _ = permutation_cluster_1samp_test(
        diff_array_clean, threshold=None, tail=0,
        adjacency=coo_matrix(adj_matrix[np.ix_(retained_rois, retained_rois)]).tocsr(),
        n_permutations=n_permutations,
        out_type="indices", seed=42
    )

    for i, (roi_idxs,) in enumerate(clusters):
        p_val = cluster_pv[i]
        if p_val < alpha_threshold:
            for local_idx in roi_idxs:
                roi = retained_rois[local_idx]
                t_obs_per_param_roi[(param, roi)] = t_obs[local_idx]


for row in results_df_rows:
    key = (row["parameter"], row["roi"])
    if key in t_obs_per_param_roi:
        significant_roi_rows.append({
            "parameter": row["parameter"],
            "roi": row["roi"],
            "t_value": row["t_value"],
            "p_value_uncorrected": row["p_value"],
            "t_obs_cluster": t_obs_per_param_roi[key],
            "p_value_cluster": 10**(-corrected_logp_map[key]) if key in corrected_logp_map else None
        })

significant_df = pd.DataFrame(significant_roi_rows)

print("\nSignificant ROI Summary")
print(significant_df)

total_clusters = results_df["n_clusters"].sum()
total_sig_clusters = results_df["n_sig_clusters"].sum()

print(f"\nTotal clusters found across all parameters: {total_clusters}")
print(f"Significant clusters (p < {alpha_threshold}): {total_sig_clusters}")


for _, cluster_row in cluster_table_df[cluster_table_df["p_value"] < alpha_threshold].iterrows():
    param = cluster_row["parameter"]
    p_val_cluster = cluster_row["p_value"]
    for roi in cluster_row["rois"]:

        match = next((r for r in results_df_rows if r["parameter"] == param and r["roi"] == roi), None)
        if match:
            significant_roi_rows.append({
                "parameter": param,
                "roi": roi,
                "t_value": match["t_value"],
                "p_value_uncorrected": match["p_value"],
                "t_obs_cluster": match["t_value"], 
                "p_value_cluster": p_val_cluster
            })


sig_clusters_df = cluster_table_df[cluster_table_df["p_value"] < alpha_threshold]
print("\nBreakdown of significant clusters (ROIs per cluster):")
for i, row in sig_clusters_df.iterrows():
    print(f"- Param: {row['parameter']}, Cluster {row['cluster_id']}: {len(row['rois'])} ROI(s)")
```
